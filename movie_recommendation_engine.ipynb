# Movie Recommendation Engine

```python
import pandas as pd
```

```python
df1=pd.read_csv("movies.csv")
df1.head()
```

```python
df2=pd.read_csv("ratings.csv")
df2.head()
```

```python
df3=pd.merge(df1,df2).drop(['genres','timestamp'],axis=1)
df3.head()
```

```python
userRatings = df3.pivot_table(index=['userId'],columns=['title'],values='rating')
userRatings.head()
print("Before: ",userRatings.shape)
userRatings = userRatings.dropna(thresh=10, axis=1).fillna(0,axis=1)

#userRatings.fillna(0, inplace=True)
print("After: ",userRatings.shape)
userRatings.head()
```

```python
corrMatrix = userRatings.corr(method='pearson')
corrMatrix
```

```python
for i in corrMatrix.index:
    print(i)
```

```python
def get_similar(movie_name,rating):
    similar_ratings = corrMatrix[movie_name]*(rating-2.5)
    similar_ratings = similar_ratings.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_ratings
```

```python
movies = [("(500) Days of Summer (2009)",5),("Alice in Wonderland (2010)",3),("Aliens (1986)",1),("2001: A Space Odyssey (1968)",2)]
similar_movies = pd.DataFrame()
for movie,rating in movies:
    similar_movies = similar_movies.append(get_similar(movie,rating),ignore_index = True)
```

```python
similar_movies.sum().sort_values(ascending=False).head(10)
```
